```python
#create a function that returns the dictionary for the album

def find_by_name(album):
    for item in rs_dict:
        if item['album'] == album:
            return item
        else:
            return ('WRONG')

find_by_name('Velvet Underground')
```

```python
#create a function that return album dictionary by rank
def find_by_rank(rank):
    for item in rs_dict:
        if item['number'] == rank:
            return item
        else:
            if rank > '500':
                return False
        
        
find_by_rank('499')
```

```python
def find_by_year(year):
    album_year = []
    for item in rs_dict:
        if item['year'] == year:
            album_year.append(item['album'])
    return album_year

find_by_year('1980')        
```

```python
#create a function that returns a list of albums that are released between 2 given years
def find_by_years(year1, year2):
    album_list = []
    for item in rs_dict:
        if item['year'] >= year1 and item['year'] <= year2:
            album_list.append(item['album'])
    return album_list

find_by_years('1970', '1980')
    
```

```python
#find out how many albums each artist has
from collections import defaultdict


artist_list = all_artists()

d = defaultdict(int)
for artist in artist_list:
    d[artist] += 1

```

```python
#create a dictionary of artist to number of albums
artist_dict = dict(d)
artist_dict
```

```python
#Artists with the most albums - Returns the artist with the highest amount of albums on the list of top albums
def artist_most_albums():
    artist_list = []
    
    list_of_val = list(artist_dict.values())
    list_of_keys = list(artist_dict.keys())
    
    for idx, val in enumerate(list_of_val):
        if val == max(list_of_val):
            artist_list.append(list_of_keys[idx])
    return artist_list
    

artist_most_albums()
```

```python
#Histogram of albums by decade - 
#Returns a histogram with each decade pointing to the number of albums released during that decade.
import matplotlib.pyplot as plt

#create a list of years
x = []

for item in rs_dict:
    x.append(item['year'])

#convert years to inegers
years = [int(year) for year in x]

plt.hist(years, bins=7, edgecolor='black', rwidth=.5, range = [1950, 2020])
plt.xlabel('Decades')
plt.ylabel('Number of Albums')
plt.title('Albums By Decade')

plt.Show()
```
